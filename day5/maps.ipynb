```python
# Imports at the top of the notebook

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# Get linearly spaced points:
x = np.linspace(0, 10, 100)
y_cos = np.cos(x)

plt.plot(x, y_cos)
```

```python
# Follows matlab syntax- 'r-x' means red, lines, x's on points
plt.plot(x, y_cos, 'r-x')
```

```python
# Using parameter names
plt.plot(x, y_cos, color='red', linestyle='-')
```

```python
# Plot both at once
y_sin = np.sin(x)
plt.plot(x, y_cos, x, y_sin)
```

```python
# Plot both at once
y_sin = np.sin(x)
plt.plot(x, y_cos)
plt.plot(x, y_sin)
```

```python
'''
Plots for EPA fuel economy data
'''

cars = pd.read_csv('vehicles.csv')

g = cars.groupby(['make', 'year'])
brand_mpg = g['city_mpg'].mean()
brand_mpg = brand_mpg.reset_index()
    
# Pivot
mpg_matrix = brand_mpg.pivot('year', 'make', 'city_mpg')

# Subset some brands
brands = ['Toyota', 'Honda', 'Hyundai', 'Ford', 'Chevrolet', 'Ferrari', 'Lamborghini', 'BMW', 'Mercedes-Benz']
mpg_matrix = mpg_matrix[brands]

# Plot
sns.heatmap(mpg_matrix)
```

```python
# Use base matplotlib technique:

# Plot
fig, ax = plt.subplots(1, 1)
ax.set_xlabel('Brand')
ax.set_ylabel('Year')
ax.set_title('MPG')
sns.heatmap(mpg_matrix, ax=ax)

# Don't need this running interactively in the notebook
#fig.show()
```
