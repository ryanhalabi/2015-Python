```python
lastweek = ', '.join(feedback['finalweek'])
count = collections.Counter(lastweek.split('), '))

common = count.most_common()

# Create a bar plot
plt.bar(left=np.arange(len(common)), height = [x[1] for x in common])

# Add labels
for i, tup in enumerate(common):
    plt.text(i, tup[1], tup[0])
```

```python
for i, colname in enumerate(['pace', 'recommend', 'relevance', 'learn', 'teachstyle']):
    plt.figure(i)
    feedback[colname].plot(kind='hist')
    plt.title(colname)
```
